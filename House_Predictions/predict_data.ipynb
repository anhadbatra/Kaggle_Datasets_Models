```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

df = pd.read_csv("C:/Users/itisa/Documents/Data/Kaggle_competion/Kaggle_Datasets_Models/House_Predictions/train.csv")
```

```python
df['Street'] = df['Street'].map({'Pave':0,'Grvl':1})
df['SaleCondition'] = df['SaleCondition'].map({'Normal':1,'Abnorml':2,'Partial':3,'Adjland':4,'Alloca':5,'Family':6})
df['SaleType'] = df['SaleType'].map({'WD': 0, 'New': 1, 'COD': 2, 'ConLD': 3, 'ConLI': 4, 'CWD': 5, 'ConLw': 6, 'Con': 7, 'Oth': 8})
df['PavedDrive'] = df['PavedDrive'].map({'Y': 1, 'N': 0, 'P': 2})
df['LotShape'] = df['LotShape'].map({'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3})
df['MSZoning'] = df['MSZoning'].map({'RL': 0, 'RM': 1, 'C (all)': 2, 'FV': 3, 'RH': 4})
df['Utilities'] = df['Utilities'].map({'AllPub': 0, 'NoSeWa': 1})
df['LandSlope'] = df['LandSlope'].map({'Gtl': 0, 'Mod': 1, 'Sev': 2})
df['LandContour'] = df['LandContour'].map({'Lvl': 0, 'Bnk': 1, 'Low': 2, 'HLS': 3})
df['LotConfig'] = df['LotConfig'].map({'Inside': 0, 'Corner': 1, 'CulDSac': 2, 'FR2': 3, 'FR3': 4})
df['Condition1'] = df['Condition1'].map({'Norm': 0, 'Feedr': 1, 'PosN': 2, 'Artery': 3, 'RRAe': 4, 'RRNn': 5, 'RRAn': 6, 'PosA': 7, 'RRNe': 8})
df['BldgType'] = df['BldgType'].map({'1Fam': 0, '2fmCon': 1, 'Duplex': 2, 'TwnhsE': 3, 'Twnhs': 4})
df['HouseStyle'] = df['HouseStyle'].map({
    '2Story': 0,
    '1Story': 1,
    '1.5Fin': 2,
    '1.5Unf': 3,
    'SFoyer': 4,
    'SLvl': 5,
    '2.5Unf': 6,
    '2.5Fin': 7
})
df['RoofStyle'] = df['RoofStyle'].map({
    'Gable': 0,
    'Hip': 1,
    'Gambrel': 2,
    'Mansard': 3,
    'Flat': 4,
    'Shed': 5
})

df.drop(columns=['PavedDrive','EnclosedPorch','Alley','PoolQC','Fence','MiscFeature','MiscVal','MoSold','Neighborhood','RoofMatl','Exterior1st','Exterior2nd'], inplace=True)
df.dropna()
print(df)
```

```python
X=df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

```python
param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}

rf = RandomForestClassifier()
rand_search = RandomizedSearchCV(rf, 
                                 param_distributions = param_dist, 
                                 n_iter=5, 
                                 cv=5)
rand_search.fit(X,y)
```

```python
test_df = pd.read_csv("test.csv")
test_df['Street'] = test_df['Street'].map({'Pave':0,'Grvl':1})
test_df['SaleCondition'] = test_df['SaleCondition'].map({'Normal':1,'Abnorml':2,'Partial':3,'Adjland':4,'Alloca':5,'Family':6})
test_df['SaleType'] = test_df['SaleType'].map({'WD': 0, 'New': 1, 'COD': 2, 'ConLD': 3, 'ConLI': 4, 'CWD': 5, 'ConLw': 6, 'Con': 7, 'Oth': 8})
test_df['LotShape'] = test_df['LotShape'].map({'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3})
test_df['MSZoning'] = test_df['MSZoning'].map({'RL': 0, 'RM': 1, 'C (all)': 2, 'FV': 3, 'RH': 4})
test_df['Utilities'] = test_df['Utilities'].map({'AllPub': 0, 'NoSeWa': 1})
test_df['LandSlope'] = test_df['LandSlope'].map({'Gtl': 0, 'Mod': 1, 'Sev': 2})
test_df['LandContour'] = test_df['LandContour'].map({'Lvl': 0, 'Bnk': 1, 'Low': 2, 'HLS': 3})
test_df['LotConfig'] = test_df['LotConfig'].map({'Inside': 0, 'Corner': 1, 'CulDSac': 2, 'FR2': 3, 'FR3': 4})
test_df['Condition1'] = test_df['Condition1'].map({'Norm': 0, 'Feedr': 1, 'PosN': 2, 'Artery': 3, 'RRAe': 4, 'RRNn': 5, 'RRAn': 6, 'PosA': 7, 'RRNe': 8})
test_df['BldgType'] = test_df['BldgType'].map({'1Fam': 0, '2fmCon': 1, 'Duplex': 2, 'TwnhsE': 3, 'Twnhs': 4})
test_df['HouseStyle'] = test_df['HouseStyle'].map({
    '2Story': 0,
    '1Story': 1,
    '1.5Fin': 2,
    '1.5Unf': 3,
    'SFoyer': 4,
    'SLvl': 5,
    '2.5Unf': 6,
    '2.5Fin': 7
})
test_df['RoofStyle'] = test_df['RoofStyle'].map({
    'Gable': 0,
    'Hip': 1,
    'Gambrel': 2,
    'Mansard': 3,
    'Flat': 4,
    'Shed': 5
})

test_df.drop(columns=['EnclosedPorch','Alley','PoolQC','Fence','MiscFeature','MiscVal','MoSold','Neighborhood','RoofMatl','Exterior1st','Exterior2nd'], inplace=True)
test_df.dropna()
print(test_df)
```

```python
X = test_df
y_pred = rand_search.predict(X)

output = pd.DataFrame({'id': test_df['id'], 'SalePrice': y_pred})
output.to_csv('submission.csv', index=False)
```
